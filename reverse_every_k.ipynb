```python
# Given a linked list, write a function to reverse every k nodes
```

```python
def reverse_list(x):
    current_node = x.head
    previous_node = None
    next_node = None
    while current_node:
        next_node = current_node.next_node
        current_node.next_node = previous_node
        previous_node = current_node
        current_node = next_node
    x.head = previous_node

def reverse_list_k(x, k):
    stack = []
    new_head = None
    tail = None
    while not x.is_empty():
        # push k elements on stack
        for i in range(k):
            node = x.pop()
            if node:
                stack.append(node)
        
        #pop top of stack, either it's the new head or it should be merged with the current tail
        node = stack.pop()
        if new_head is None:
            new_head = node
        else:
            tail.next_node = node
            
        while len(stack) > 0:
            node.next_node = stack.pop()
            node = node.next_node
        tail = node
    
    # nil out last tail pointer
    tail.next_node = None
    # reset new head
    x.head = new_head

class Node:
    def __init__(self, data):
        self.data = data
        self.next_node = None


class LinkedList:
    def __init__(self, head=None):
        self.head = head

    def insert(self, data):
        node = Node(data)
        if self.head is None:
            self.head = node
        else:
            n = self.head
            while n.next_node:
                n = n.next_node
            n.next_node = node

    def pop(self):
        h = self.head
        if h:
            self.head = h.next_node
        else:
            self.head = None
        return h

    def is_empty(self):
        return self.head is None

    def print_list(self):
        n = self.head
        while n:
            print(n.data),
            n = n.next_node
```

```python
x = LinkedList()
for i in range(1, 17):
    x.insert(i)

x.print_list()
reverse_list_k(x, 3)
print('reversed')
x.print_list()
```
