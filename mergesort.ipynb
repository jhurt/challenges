```python
# two distinct processes that make up its implementation. 
# First, the list is split into halves. divide a list in half logn times where n is the length of the list. 
# The second process is the merge. Each item in the list will eventually be processed and placed on the sorted list. 
# So the merge operation which results in a list of size n requires n operations. The result of this analysis is that logn splits, each of which costs n for a total of nlogn operations. A merge sort is an O(nlogn) algorithm.
```

```python
import random

my_list = []
for i in range(100):
    my_list.append(random.randint(1, 1000))
my_list
```

```python
def merge_sort(list_in):
    if len(list_in) > 1:
        middle_index = len(list_in) // 2
        left_half = list_in[:middle_index]
        right_half = list_in[middle_index:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = 0
        j = 0
        k = 0
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                list_in[k] = left_half[i]
                i += 1
            else:
                list_in[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            list_in[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            list_in[k] = right_half[j]
            j += 1
            k += 1

merge_sort(my_list)
```
